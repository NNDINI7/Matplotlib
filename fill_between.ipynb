```python
#Basic Syntax  
# simple line plot 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y,color="r")
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
#Basic Syntax  
# simple line plot 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

#to fill the area under the line
plt.fill_between(x, y, color="skyblue")

plt.plot(x, y,color="r")
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python


import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

#to fill the area under the line
plt.fill_between(x=[2,4], y1=2, y2=4, color="skyblue")
#y1 and y2 are the y-coordinates of the fill area
#x=[2,4] specifies the x-coordinates where the fill should occur

plt.plot(x, y,color="r")
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
# 

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

#to fill the area under the line
plt.fill_between(x,y, color="skyblue", where=(x >= 2) & (x <= 4)) 

#where specifies the condition for filling



plt.plot(x, y,color="r")
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```
