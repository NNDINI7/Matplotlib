```python
number = [55, 12, 41, 29, 38, 41, 45, 43 ,54 ,13 ,28, 49 ,18 ,22 ,59 ,56 ,42 ,54 ,19 ,54 ,56 ,51 ,15 ,35,
 58 ,18, 39, 36, 17, 32, 14, 28, 52, 25, 25, 55, 14, 13, 58, 53, 56, 13, 42, 15, 25 ,33 ,18, 44,
 35, 38]
plt.hist(number,color = "purple",bins = 10 ,edgecolor = "yellow",cumulative=True)
# This adds a color to the plot
#bins is used to change the number of bins in the histogram.
#edgecolor is used to change the color of the edges of the bars.
#cumulative=-1 is used to plot the histogram in reverse order.
#cumulative=True is used to plot the cumulative histogram.


plt.title("Histogram", fontsize=20)
plt.xlabel("Number", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.show()
```

```python
number = [55, 12, 41, 29, 38, 41, 45, 43 ,54 ,13 ,28, 49 ,18 ,22 ,59 ,56 ,42 ,54 ,19 ,54 ,56 ,51 ,15 ,35,
 58 ,18, 39, 36, 17, 32, 14, 28, 52, 25, 25, 55, 14, 13, 58, 53, 56, 13, 42, 15, 25 ,33 ,18, 44,
 35, 38]
plt.hist(number,color = "purple",edgecolor = "yellow" , orientation='vertical', rwidth=0.8)
# histtype='stepfilled' is used to fill the histogram bars with color.
# This will fill the histogram bars with color.
#step is used to create a step histogram.
#barstacked is used to create a stacked histogram.

#orientation='horizontal' is used to change the orientation of the histogram to horizontal.
# This will change the orientation of the histogram to horizontal.


plt.title("Histogram", fontsize=20)
plt.xlabel("Number", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.show()
```

```python
number = [55, 12, 41, 29, 38, 41, 45, 43 ,54 ,13 ,28, 49 ,18 ,22 ,59 ,56 ,42 ,54 ,19 ,54 ,56 ,51 ,15 ,35,
 58 ,18, 39, 36, 17, 32, 14, 28, 52, 25, 25, 55, 14, 13, 58, 53, 56, 13, 42, 15, 25 ,33 ,18, 44,
 35, 38]
plt.hist(number,color = "purple",edgecolor = "yellow", log=True, label ='Log Scale')
# log=True is used to plot the histogram in logarithmic scale.
# This will plot the histogram in logarithmic scale.


plt.title("Histogram", fontsize=20)
plt.xlabel("Number", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.legend()
#legend() is used to add a legend to the plot.
plt.axvline(45,color = 'red')
#axvline() is used to add a vertical line to the plot.
# This will add a vertical line at x=45.
plt.grid(True)
#grid() is used to add a grid to the plot.


plt.show()
```
