writing different types of text in matplotlib

```python
# types of text in matplotlib

# text = add text at an arbitrary position of the axes
#annotate = add an annotation with an arrow, at an arbitrary position of the axes
# xlabel = add a label to the x-axis
# ylabel = add a label to the y-axis
# title = add a title to the axes
```

```python
import matplotlib.pyplot as plt
#linear plot

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.plot(x, y)
plt.title("Linear Plot")
plt.show()

```

```python
import matplotlib.pyplot as plt
#linear plot

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.plot(x, y)

plt.title("Linear Plot with text " , fontsize = 20 )

plt.xlabel("DAYS", fontsize=15)
plt.ylabel("PYTHON", fontsize=15)
plt.text(2, 3, "python", fontsize=12, color='red')  # Add text at a specific position
#text is used to add text at a specific position in the axes

plt.show()

```

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.plot(x, y)

plt.title("Linear Plot with text " , fontsize = 20 )

plt.xlabel("DAYS", fontsize=15)
plt.ylabel("PYTHON", fontsize=15)
plt.text(2, 3, "python", fontsize=12, style='italic',bbox={"facecolor":"orange"})  # Add italic text at a specific position

plt.show()

```

```python
import matplotlib.pyplot as plt
#linear plot

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.plot(x, y)

plt.title("Linear Plot with text " , fontsize = 20 )

plt.xlabel("DAYS", fontsize=15)
plt.ylabel("PYTHON", fontsize=15)
plt.text(2, 3, "python", fontsize=12, color='red')  

plt.annotate('This is a point', xy=(2, 3), xytext=(3, 4) , arrowprops=dict(facecolor = "black" , shrink = 100 ))  # Add an
# annotation with an arrow at a specific position
#annotate is used to add an annotation with an arrow, at an arbitrary position of the axes
# arrowprops is used to set the properties of the arrow
# xy is the position of the text
# xytext is the position of the arrow
# shrink is used to set the size of the arrow

plt.legend(['Line 1'], loc=9 ,facecolor ="red" ,edgecolor = "yellow" , framealpha = 0.5)  # Add a legend to the plot
plt.show()

```
