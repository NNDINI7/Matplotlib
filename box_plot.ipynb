```python

#Basic syntax 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,6,7]


plt.boxplot(x)
plt.show()
```

```python

#Basic syntax 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,6,7]


plt.boxplot(x, notch = True)
#notch shows the confidence interval around the median.

plt.show()
```

```python

#Basic syntax 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,6,7]


plt.boxplot(x, vert=False)
#vert=False makes the box plot horizontal

plt.show()
```

```python

#Basic syntax 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,6,7]


plt.boxplot(x, widths=0.5)
#widths controls the width of the box plot


plt.show()
```

```python

#Basic syntax 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,6,7]


plt.boxplot(x, labels=["python"] , patch_artist=True)
#labels sets the labels for the box plot. it should be a list of strings with the same length as the data.
#patch_artist=True fills the box with color

plt.show()
```

```python

#Basic syntax 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,6,7]


plt.boxplot(x, showmeans=True)
#showmeans=True shows the mean of the data as a point on the box plot

plt.show()
```

```python

#Basic syntax 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,6,7,15,20]


plt.boxplot(x, showmeans=True,sym="r+" ,
            boxprops=dict(color='b'),
              whiskerprops=dict(color='g'),
                capprops=dict(color='r'), 
                medianprops=dict(color='b'),flierprops=dict(marker='o', color='y'))
#showmeans=True shows the mean of the data as a point on the box plot
#boxprops controls the properties of the box. facecolor sets the fill color, color sets the border color.
#whiskerprops controls the properties of the whiskers. color sets the color of the whiskers.
#capprops controls the properties of the caps. color sets the color of the caps.
#medianprops controls the properties of the median line. color sets the color of the median line.
#flierprops controls the properties of the outliers. marker sets the marker style, color sets the color of the outliers.


#whis controls the length of the whiskers. 1.5 is the default value, it means that the whiskers will extend to 1.5 times the interquartile range (IQR) from the quartiles.
#sym controls the symbol used to represent the outliers. 'r+' means red plus sign.

plt.show()
```

```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,6,7]
x1= [1, 2, 3, 4, 5,6,7,15]
y = [x, x1]


plt.boxplot(y, labels=["Python", "Java"], patch_artist=True)
#labels sets the labels for the box plot. it should be a list of strings with the same length as the data.
#patch_artist=True fills the box with color


plt.show()
```
