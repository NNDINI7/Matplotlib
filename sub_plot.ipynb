```python
#use of subplot
#syntax : plt.subplot(nrows, ncols, index)
#nrows: number of rows in the subplot grid
#ncols: number of columns in the subplot grid
#index: index of the subplot to be created (1-based index)


import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1,2,3,4,5]

plt.subplot(2, 2, 1)  # 2 rows, 2 columns, first subplot
plt.plot(x, y, color="red")

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, second subplot
plt.pie([1], colors ="red")

x1 = [10, 20, 30, 40, 50]
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, third subplot
plt.pie(x)

x2 = ["a", "b", "c", "d", "e"]
y2 =[10, 20, 30, 40, 50]
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, fourth subplot
plt.bar(x2, y2, color="green")

plt.show()
```
