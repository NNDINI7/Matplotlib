```python
#basic syntax :
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,5,4,3,6]
plt.scatter(day, number)
plt.title("Scatter ",fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("Number", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,5,4,3,6]

color = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink']
size = [50, 100, 150, 200, 250, 300, 350]
#size is used to change the size of the points
plt.scatter(day, number,c = color, marker='o',s=size,edgecolors='black',linewidths=2)
#color is used to change the color of the points
#marker is used to change the shape of the points

plt.title("Scatter ",fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("Number", fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,5,4,3,6]

color = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink']
size = [50, 100, 150, 200, 250, 300, 350]
#size is used to change the size of the points
plt.scatter(day, number,c = color, marker='*',s=size, alpha=0.5)
#alpha is used to change the transparency of the points. is is used to blurs the graph ponits.
#color is used to change the color of the points
#marker is used to change the shape of the points

plt.title("Scatter ",fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("Number", fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,5,4,3,6]
number2 = [3, 4, 2, 6, 5, 4, 7]

color = [10, 20, 30, 40, 50, 60, 70]
#we asign color as a list of numbers to show the difference in color.
size = [50, 100, 150, 200, 250, 300, 350]
#size is used to change the size of the points

plt.scatter(day, number,c = color, marker='*',s=size, alpha=0.5,cmap="viridis")
plt.scatter(day, number2 ,color = "r", marker='*',s=size)

plt.colorbar()  # This adds a color bar to the plot
#cmap is used to change the color map of the points.
#alpha is used to change the transparency of the points. is is used to blurs the graph ponits.
#color is used to change the color of the points
#marker is used to change the shape of the points

plt.title("Scatter ",fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("Number", fontsize=15)

plt.show()
```
