```python
import matplotlib.pyplot as plt

x= [1, 2, 3, 4, 5]
y= [2, 3, 5, 7, 3]

plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x= [1, 2, 3, 4, 5]
y= [2, 3, 5, 7, 3]

plt.stem(x,y, linefmt='--', markerfmt='ro',bottom=0, basefmt='g', label='Stem Plot')

# Customizing the line and marker formats
#linefmt: format of the stem lines
#markerfmt: format of the markers
#ro: red circle markers
#bottom: baseline for the stems. if bottom is 9 then the stems will start from 9,it will reverse the y-axis
#basefmt: format of the baseline

plt.title('Stem Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x= [1, 2, 3, 4, 5]
y= [2, 3, 5, 7, 3]

plt.stem(x,y, orientation= "horizontal",linefmt='--', markerfmt='ro',bottom=0, basefmt='g', label='Stem Plot')

# orientation: 'horizontal' makes the stems horizontal

plt.title('Stem Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```
