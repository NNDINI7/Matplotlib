```python
# Basic syntax
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
plt.pie(x, colors=['red', 'blue', 'green', 'yellow', 'purple'])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = ['c', 'java', 'python', 'javascript', 'ruby']
# here y is the labels for the pie chart
plt.pie(x, labels=y, colors=['red', 'blue', 'green', 'yellow', 'purple'])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = ['c', 'java', 'python', 'javascript', 'ruby']
# here y is the labels for the pie chart

ex=[0.2, 0, 0, 0.3, 0]  # explode the first slice

plt.pie(x, labels=y, colors=['red', 'blue', 'green', 'yellow', 'purple'], explode=ex)
# explode is used to separate the first slice from the rest

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = ['c', 'java', 'python', 'javascript', 'ruby']
# here y is the labels for the pie chart

plt.pie(x, labels=y, colors=['red', 'blue', 'green', 'yellow', 'purple'],autopct='%0.1f%%' , shadow=True , radius=1.2, labeldistance=1.2)
# autopct is used to display the percentage of each slice
#format is used to format the percentage. 

# shadow is used to add a shadow effect to the pie chart
# radius is used to change the size of the pie chart
# labeldistance is used to change the distance of the labels from the center of the pie chart




plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = ['c', 'java', 'python', 'javascript', 'ruby']
# here y is the labels for the pie chart
plt.pie(x, labels=y,  startangle=90, counterclock=False, wedgeprops={'edgecolor': 'black',"linewidth":5 } , textprops={'fontsize': 14})
#linewidth is used to set the width of the edge of the pie chart
# textprops is used to set the properties of the text in the pie chart
# startangle is used to rotate the start of the pie chart
# counterclock is used to change the direction of the pie chart
# wedgeprops is used to set the properties of the wedges

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = ['c', 'java', 'python', 'javascript', 'ruby']
# here y is the labels for the pie chart
plt.pie(x, labels=y,rotatelabels=True )
# rotatelabels is used to rotate the labels of the pie chart

plt.title("Pie Chart", fontsize=20)
plt.legend(loc =2)
# legend() is used to add a legend to the plot.
plt.axis('equal')
# axis('equal') is used to make the pie chart a circle
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
x1 = [2, 3, 4, 5, 6]
# here x1 is the second pie chart data
y = ['c', 'java', 'python', 'javascript', 'ruby']
# here y is the labels for the pie chart
plt.pie(x, labels=y,radius = 1.5, colors=['red', 'blue', 'green', 'yellow', 'purple'])
c = plt.Circle(xy =(0, 0),radius = 1, facecolor='white')
plt.gca().add_artist(c)
# Adding a white circle in the center to create a donut chart effect


plt.show()
```
