```python
#linear plot 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.plot(x, y, color="red")
plt.savefig("linear_plot.png")  # Save the figure as a PNG file in the current directory
plt.show()

```

```python
#use of parameters 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 6, 7, 4, 5]

plt.plot(x, y, color="red")
plt.savefig("linear1.png" ,dpi=2000) 
#dpi is used to set the resolution of the saved figure

plt.show()

```

```python
#use of parameters 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 6, 7, 4, 5]

plt.plot(x, y, color="red")
plt.savefig("linear2.png" ,facecolor='blue', edgecolor='black', dpi=2000) 
#facecolor is used to set the background color of the saved figure
#edgecolor is used to set the border color of the saved figure

plt.show()

```

```python
#use of parameters 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 6, 7, 4, 5]

plt.plot(x, y, color="red")
plt.savefig("linear3.pdf" ,facecolor= "blue",transparent = True)

plt.show()

```

```python
#use of parameters 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 6, 7, 4, 5]

plt.plot(x, y, color="red")
plt.savefig("linear4.png" ,facecolor='blue', edgecolor='black', bbox_inches='tight', dpi=2000)
#bbox_inches='tight' is used to adjust the bounding box of the saved figure
#transparent=True is used to save the figure with a transparent background

#dpi is used to set the resolution of the saved figure

plt.show()

```
