```python
import matplotlib.pyplot as plt 
x = ["python","c","R","java"]
y = [85,70,80,82]

plt.xlabel("Languages", fontsize=14)
plt.ylabel("Popularity", fontsize=14)
plt.title("Programming Language Popularity", fontsize=16)
#fontsize is used to increase the size of the text.

c=['blue', 'green', 'red', 'orange']
plt.bar(x,y, color=c, width=0.4, edgecolor='yellow',align='center',linestyle=':', alpha=0.7)
#align is used to align the bars to the center of the x-ticks
#linestyle is used to set the style of the edge of the bars
#alpha is used to set the transparency of the bars

plt.show()
```

```python
import matplotlib.pyplot as plt 
x = ["python","c","R","java"]
y = [85,70,80,82]
z= [20,30,40,50]

plt.xlabel("Languages", fontsize=14)
plt.ylabel("Popularity", fontsize=14)
plt.title("Programming Language Popularity", fontsize=16)

plt.bar(x,y, color="purple", width=0.4, edgecolor='black',align='center',label="Popularity")
plt.bar(x,z, color="blue", width=0.4, edgecolor='black',align='center',label="Popularity1")
#we use 2 bars to represent 2 different datasets

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
x = ["python","c","R","java"]
y = [85,70,80,82]
z= [20,30,40,50]
# z is another dataset to be plotted alongside y.

width = 0.4

# we use numpy to create an array of positions for the bars.
p = np.arange(len(x))  # the label locations
p1 = [x + width for x in p]  # the label locations for the second bar

plt.xlabel("Languages", fontsize=14)
plt.ylabel("Popularity", fontsize=14)
plt.title("Programming Language Popularity", fontsize=16)

plt.bar(p,y, color="red", width=0.4, edgecolor='black',align='center',label="Popularity")
plt.bar(p1,z, color="yellow", width=0.4, edgecolor='black',align='center',label="Popularity1")

plt.xticks(p+width/2,x,rotation =45)  # Set the x-ticks to the center of the bars. we divide by 2 to center the ticks between the two bars.
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = ["python","c","R","java"]
y = [85,70,80,82]
z= [20,30,40,50]

plt.xlabel("Languages", fontsize=14)
plt.ylabel("Popularity", fontsize=14)
plt.title("Programming Language Popularity", fontsize=16)

#barh is used to create horizontal bar charts
plt.barh(x,y, color="purple", label="Popularity")
plt.barh(x,z, color="blue",label="Popularity1")


plt.legend()
plt.show()
```
