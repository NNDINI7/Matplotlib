```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [7,4,2,3,5]
plt.plot(x, y)
plt.show()
```

```python
#make use of xticks and yticks to set the ticks on the x and y axis

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [7,4,2,3,5]
plt.plot(x, y)

#set the numbers on axis accordingly 

plt.xticks(x)
plt.yticks(x)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [7,4,2,3,5]

plt.plot(x, y)

plt.xticks(x, labels=['one', 'two', 'three', 'four', 'five'])
plt.yticks(y, labels=['seven', 'four', 'two', 'three', 'five'])
#labels are used to set the labels of the ticks on the x and y axis

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [7,4,2,3,5]

plt.plot(x, y)

plt.xlim(0, 6)  # Set x-axis limits
plt.ylim(0, 8)  # Set y-axis limits
#xlim and ylim are used to set the limits of the x and y axis

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [3,1,5,2]

plt.plot(x, y)
plt.axis([0,10,0,7])  # Set x and y axis limits in one go
#axis is used to set the limits of the x and y axis in one go

plt.show()
```
