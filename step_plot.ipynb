```python


import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 3, 4, 5, 6, 7]

plt.step(x,y)

plt.title("Step Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

plt.show()
```

```python
#step plot function parameters :


import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 3, 4, 5, 6, 7]

plt.step(x,y,color="r", marker='o',ms = 10, mfc ="y", label = "python")

#color : color of the step line
#marker : marker style for the data points
#ms : marker size
#mfc : marker face color
#label : label for the step line

plt.title("Step Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

plt.legend(loc = 2)

plt.show()
```
